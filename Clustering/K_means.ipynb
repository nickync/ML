# Import models

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import dataset

```python
dataset = pd.read_csv('winequality-red.csv')
X = dataset.iloc[:, 9:11].values
```

# Using elbow method to find the optimal clusters

```python
from sklearn.cluster import KMeans

wcss = []

for i in range (1, 11):
  k_means = KMeans(n_clusters = i, init = 'k-means++')
  k_means.fit(X)
  wcss.append(k_means.inertia_)

plt.plot(range(1, 11), wcss)
```

# Train the Kmeans model 

```python
k_means = KMeans(n_clusters=4, init='k-means++')
y_kmeans = k_means.fit_predict(X)
print(y_kmeans)
```

# Plot the clusters

```python
plt.scatter(X[:,0], X[:,1], c = y_kmeans, cmap = 'rainbow')
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], marker='v', s = 100, c = 'magenta')
plt.title('......')
plt.xlabel('sulpates')
plt.ylabel('achohol')
plt.show()
```
