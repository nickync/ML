# import library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# import dataset

```python
dataset = pd.read_csv('marketbasket.csv')
```

# preprocess dataset

```python
dataset = dataset.astype('str')

for i in range(0, 1361):
    for j in range(0, 255):
        if dataset.values[i,j] == '0':
            dataset.values[i,j] = 'nan'
        elif dataset.values[i,j] != 0:
            dataset.values[i,j] = dataset.columns[j]
display(dataset)
```

# prepare dataset

```python
transactions = []
for i in range (0, 1361):
    transactions.append([dataset.values[i,j] for j in range (0, 255)])

print(transactions)
```

# train the apriori model

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.03, min_confidence = 0.4, min_lift = 5, min_length = 2, max_length = 2 )
```

# show results

```python
results = list(rules)
results
```

# show results in dataframe

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidences', 'lifts'])
```

```python
resultsinDataFrame.nlargest(n = 10, columns = 'lifts')
```
